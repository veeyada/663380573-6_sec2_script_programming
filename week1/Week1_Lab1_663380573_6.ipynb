2. Declare and print variables:

3. Perform simple arithmetic:


```python
print("--- ข้อ 3: การคำนวณทางคณิตศาสตร์ ---")
num1 = 15
num2 = 4
print(f"ผลบวก (Sum): {num1 + num2}")
print(f"ผลลบ (Difference): {num1 - num2}")
print(f"ผลคูณ (Product): {num1 * num2}")
print(f"ผลหาร (Division): {num1 / num2}")
print()
```

```python
name = "Alice"
age = "30"  # สังเกตว่ามีเครื่องหมายคำพูดครอบอยู่
message = name + " is " + age + " years old."
print(message)
print(type(age))
```

เหตุผล: เนื่องจากเลข 30 ถูกเขียนอยู่ในเครื่องหมายคำพูด "" Python จึงมองว่ามันเป็นข้อความ (str) ไม่ใช่ตัวเลขจำนวนเต็ม การใช้เครื่องหมาย + จึงเป็นการนำข้อความมาต่อกัน (Concatenation) ได้ตามปกติโดยไม่มีข้อผิดพลาดครับ

```python
is_sunny = True
is_raining = False
print(is_sunny and is_raining)
print(type(is_sunny))
```

เหตุผล: ตัวดำเนินการตรรกศาสตร์ and (และ) จะให้ผลลัพธ์เป็น True ก็ต่อเมื่อเงื่อนไขทั้งสองฝั่งเป็นจริงทั้งคู่ แต่ในเมื่อ is_raining เป็น False ผลลัพธ์รวมจึงออกมาเป็น False และประเภทข้อมูลของตัวแปรนี้คือ Boolean (bool)

```python
num_str = input("Enter a number: ")  # สมมติว่าผู้ใช้พิมพ์เลข 10 ลงไป
result = int(num_str) + 5
print(result)
```

เหตุผล (เกิด Error!): ฟังก์ชัน input() จะส่งค่ากลับมาเป็นข้อความ (str) เสมอ แม้ว่าผู้ใช้จะพิมพ์ตัวเลขลงไปก็ตาม ดังนั้นเมื่อเราสั่งเอาตัวแปร num_str (ที่เป็นข้อความ) ไปบวกกับเลข 5 (ที่เป็น integer) Python จะไม่เข้าใจและพังทันที วิธีแก้คือต้องแปลงค่าให้เป็นตัวเลขก่อนด้วยคำสั่ง int(num_str) เหมือนใน Lab ข้อ 5 ครับ
