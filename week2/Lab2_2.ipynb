2. ลำดับตรรกะการทำงาน (Logic Flow with IF-ELIF-ELSE)
ขั้นที่ 1: รับค่าทางเลือกจากผู้เล่นเก็บไว้ในตัวแปร choice

ขั้นที่ 2: ประมวลผลเงื่อนไขเพื่อแยกจุดจบของเกม (Outcomes):

ELIF (choice == "น้ำเงิน"): * ผลลัพธ์: ประตูเปิดออกไปเจอกับดักลูกศรอาบยาพิษพุ่งเข้าใส่ตัว (GAME OVER)
