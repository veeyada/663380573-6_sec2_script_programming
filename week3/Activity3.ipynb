#Activity 3.1: Debugging Infinite Loops & Loop Logic Errors

**Buggy Snippet A: Infinite Loop**

```python
# ✅ โค้ดที่แก้ไขแล้ว
counter = 0
while counter < 5:
    print(counter)
    counter += 1  # เพิ่มค่าทีละ 1 เพื่อให้เงื่อนไขเป็นเท็จในที่สุด
```

วิธีแก้ไขที่ถูกต้อง: ต้องเพิ่มค่า counter += 1 ไว้ท้ายลูป while

**Buggy Snippet B: Incorrect Range**

```python
# ✅ โค้ดที่แก้ไขแล้ว
for i in range(1, 6): # เริ่มที่ 1 สิ้นสุดก่อนถึง 6 (ได้เลข 1-5 ตามต้องการ)
    print(i)
```

วิธีแก้ไขที่ถูกต้อง: เปลี่ยนตัวหลังในฟังก์ชัน range() ให้เป็น 6 เพื่อให้นับถึง 5

**Buggy Snippet C: Misplaced Break**

```python
# ✅ โค้ดที่แก้ไขแล้ว
for char in "hello":
    if char == 'l':
        print("Found l, stopping!")
        break  # ใส่ break ไว้ในนี้ เพื่อให้ออกจากลูปเฉพาะตอนที่เจอ 'l' เท่านั้น
    print(char)
```

วิธีแก้ไขที่ถูกต้อง: หากต้องการให้โปรแกรมพิมพ์ตัวอักษรไปเรื่อยๆ จนกว่าจะเจอตัวอักษร 'l' แล้วค่อยหยุด ต้องย้าย print(char) ไปไว้ก่อน และเอา break ใส่ไว้ข้างในบล็อกของ if

#Activity 3.2: Pattern Printing with Nested Loops

#

**1.A square of asterisks**

```python
# พิมพ์บล็อกดอกจันขนาด 5x5
size = 5
for row in range(size):
    for col in range(size):
        print("*", end="") # พิมพ์ดอกจันติดกันในแนวนอน
    print() # พิมพ์ขึ้นบรรทัดใหม่เมื่อจบแต่ละแถว
```

**2.A right-angled triangle of asterisks**

```python
# พิมพ์รูปสามเหลี่ยมมุมฉาก สูง 5 แถว
height = 5
for row in range(1, height + 1):
    for col in range(row):
        print("*", end="")
    print()
```

**3.Challenge: Inverted right-angled triangle**

```python
# พิมพ์รูปสามเหลี่ยมกลับหัว โดยนับถอยหลังจาก 5 ลงไปถึง 1
height = 5
for row in range(height, 0, -1):
    for col in range(row):
        print("*", end="")
    print()
```
