#Activity 4.1: List vs. Tuple: Scenario Matching แบบชัดเจนและตรงประเด็น

1. ข้อมูลอุณหภูมิประจำวันที่มีการบันทึกไว้ตลอดหนึ่งสัปดาห์
ควรใช้: List

เพราะว่า: อุณหภูมิในแต่ละวันเป็นข้อมูลที่มีการเปลี่ยนแปลง ดึงมาอัปเดต หรืออาจจะบันทึกเพิ่มขึ้นเรื่อย ๆ ในทุก ๆ วันที่ผ่านไป จึงต้องใช้โครงสร้างข้อมูลแบบ List ที่มีความยืดหยุ่นและสามารถแก้ไขค่าได้ (Mutable)

2. ค่าสี RGB ของเฉดสีเฉพาะเฉดหนึ่ง (เช่น สีแดงคือ (255, 0, 0))
ควรใช้: Tuple

เพราะว่า: เฉดสีเฉพาะเฉดหนึ่งจะมีค่าส่วนผสมของแม่สีคงที่ตายตัวเสมอ ไม่ควรถูกเปลี่ยนแปลงในภายหลังเพื่อป้องกันไม่ให้เฉดสีเพี้ยน จึงเหมาะกับการใช้ Tuple ที่แก้ไขค่าไม่ได้ (Immutable)

3. รายการสินค้าในรถเข็นชอปปิงที่สามารถกดเพิ่มหรือลบออกได้
ควรใช้: List

เพราะว่า: พฤติกรรมของลูกค้าในรถเข็นชอปปิงจะมีการกดเพิ่มสินค้า (ใช้คำสั่ง .append()) หรือกดลบสินค้าที่เปลี่ยนใจไม่เอาแล้วออกตลอดเวลา (ใช้คำสั่ง .remove() หรือ .pop()) ข้อมูลจึงต้องเปลี่ยนแปลงได้

4. รายชื่อวันในหนึ่งสัปดาห์เรียงตามลำดับ (จันทร์ - อาทิตย์)
ควรใช้: Tuple

เพราะว่า: วันในหนึ่งสัปดาห์มี 7 วันและมีลำดับที่แน่นอนตายตัว (ไม่มีการเพิ่มวันจันทร์ที่สอง หรือลบวันศุกร์ทิ้ง) การใช้ Tuple ช่วยการันตีว่าข้อมูลชุดนี้จะไม่ถูกเผลอแก้ไขหรือลบโดยไม่ตั้งใจระหว่างการเขียนโค้ด

5. คะแนนเก็บของนักเรียนจากการส่งงานชุดต่างๆ ซึ่งอาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลังได้
ควรใช้: List

เพราะว่า: โจทย์ระบุชัดเจนว่า "อาจมีการอัปเดตหรือแก้ไขคะแนนในภายหลัง" เช่น คุณครูตรวจคะแนนผิดแล้วมาแก้ หรือนักเรียนส่งงานชิ้นใหม่เพิ่มเข้ามา จึงจำเป็นต้องใช้คุณสมบัติการเปลี่ยนค่าได้ของ List

6. พิกัด GPS ของสถานที่สำคัญหรือจุดแลนด์มาร์กที่ตั้งอยู่กับที่
ควรใช้: Tuple

เพราะว่า: พิกัดละติจูดและลองจิจูดของสถานที่สำคัญ (เช่น โบราณสถาน หรือตึกที่เป็นแลนด์มาร์ก) เป็นค่าคงที่ทางภูมิศาสตร์ที่ตั้งอยู่กับที่ ไม่มีการขยับหรือเปลี่ยนแปลง การใช้ Tuple จึงเหมาะสม ปลอดภัย และประหยัดพื้นที่หน่วยความจำมากกว่าครับ

#Activity 4.2: Interactive To-Do List (using List Operations)

```python
# activity4_2.py - Interactive To-Do List

# 1. สร้าง List ว่างเริ่มต้นไว้ตัวหนึ่ง ชื่อว่า todo_list
todo_list = []

# 2. ใช้การวนลูป while เพื่อแสดงเมนูคำสั่งให้ผู้ใช้เลือก
while True:
    print("\n=== To-Do List Management System ===")
    print("1. เพิ่มรายการงาน (Add a task)")
    print("2. ดูรายการงานทั้งหมด (View tasks)")
    print("3. ทำเครื่องหมายว่างานเสร็จสิ้นแล้ว (Delete task)")
    print("4. ออกจากโปรแกรม (Exit)")

    # 3. ใช้คำสั่ง input() เพื่อรับค่าเมนูที่ผู้ใช้เลือก
    choice = input("กรุณาเลือกเมนู (1-4): ")

    # 4. ใช้คำสั่งจัดการ List ต่างๆ ในการบริหารจัดการข้อมูล
    if choice == '1':
        # 2.1 เพิ่มรายการงาน
        task = input("กรอกชื่องานที่คุณต้องการเพิ่ม: ")
        todo_list.append(task) # ใช้ append() เพิ่มข้อมูลต่อท้าย List
        print(f"เพิ่มงาน '{task}' เข้าสู่รายการสำเร็จแล้ว!")

    elif choice == '2':
        # 2.2 ดูรายการงานทั้งหมด
        print("\n--- รายการงานปัจจุบัน ---")
        if len(todo_list) == 0: # ใช้ len() ตรวจสอบว่ามีงานในลิสต์ไหม
            print("ไม่มีงานค้างอยู่ในรายการ (ว่างเปล่า)")
        else:
            # วนลูปเพื่อพิมพ์งานออกมาพร้อมแสดง Index (เริ่มจาก 1 เพื่อความสวยงาม)
            for index, item in enumerate(todo_list):
                print(f"{index + 1}. {item}")

    elif choice == '3':
        # 2.3 ทำเครื่องหมายว่างานเสร็จสิ้นแล้ว (ลบงานออก)
        if len(todo_list) == 0:
            print("ไม่มีรายการงานในระบบให้ลบ")
        else:
            print("\n--- รายการงานปัจจุบัน ---")
            for index, item in enumerate(todo_list):
                print(f"ดัชนี [{index}]: {item}")

            print("\nรูปแบบการลบ:")
            print("1. ลบด้วยชื่อคำสั่ง (ตามชื่องาน)")
            print("2. ลบด้วยตำแหน่ง Index (ตามตัวเลข)")
            delete_type = input("กรุณาเลือกรูปแบบการลบ (1 หรือ 2): ")

            if delete_type == '1':
                task_name = input("กรอกชื่อคำสั่ง/งานที่ต้องการลบ: ")
                if task_name in todo_list: # ตรวจสอบก่อนสั่ง remove เพื่อไม่ให้โปรแกรมพัง
                    todo_list.remove(task_name) # ใช้ remove() ลบด้วยชื่อ
                    print(f"ลบงาน '{task_name}' สำเร็จแล้ว")
                else:
                    print("ไม่พบชื่องานนี้อยู่ในรายการ")

            elif delete_type == '2':
                try:
                    idx = int(input("กรอกตัวเลขดัชนี [Index] ที่ต้องการลบ: "))
                    removed_item = todo_list.pop(idx) # ใช้ pop() ลบด้วยระบุตำแหน่ง Index
                    print(f"งาน '{removed_item}' เสร็จสิ้นและถูกลบออกแล้ว!")
                except (ValueError, IndexError):
                    print("กรอก Index ไม่ถูกต้อง หรือไม่มีตำแหน่งนั้นอยู่จริง")
            else:
                print("เลือกรูปแบบการลบไม่ถูกต้อง")

    elif choice == '4':
        # 2.4 ออกจากโปรแกรม
        print("ออกจากโปรแกรมจัดการ To-Do List เรียบร้อยแล้ว สวัสดีครับ/ค่ะ")
        break # หลุดออกจาก while loop เพื่อจบโปรแกรม

    else:
        print("ตัวเลือกไม่ถูกต้อง กรุณากรอกเฉพาะหมายเลข 1-4 เท่านั้น")
```
