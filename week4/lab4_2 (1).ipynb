# lab4_2.py - Geometric Calculations (Tuples & Immutability)

```python
import math

print("--- Lab 4.2: Geometric Calculations ---")

# 1. กำหนดพิกัดจุดในรูปแบบ Tuple
point1 = (3, 4)
point2 = (6, 8)

# เข้าถึงค่าพิกัดและพิมพ์ผลลัพธ์
print("Point 1: x = {}, y = {}".format(point1[0], point1[1]))
print("Point 2: x = {}, y = {}".format(point2[0], point2[1]))

# 2. ทดลองแก้ไขค่า (เพื่อพิสูจน์เรื่อง Immutability)
# point1[0] = 5
# 💡 เหตุผลที่แก้ไม่ได้: บรรทัดด้านบนถูกใส่ # คอมเมนต์ทิ้งไว้เนื่องจากหากรันจะเกิด "TypeError: 'tuple' object does not support item assignment"
# เพราะ Tuple เป็นโครงสร้างข้อมูลประเภท Immutable (ไม่สามารถเปลี่ยนแปลง แก้ไข เพิ่ม หรือลบสมาชิกหลังจากสร้างขึ้นมาได้)

# 3. คำนวณระยะห่างระหว่างจุด (Calculate Distance)
distance = math.sqrt((point2[0] - point1[0]) ** 2 + (point2[1] - point1[1]) ** 2)
print(f"Distance between point1 and point2: {distance}")


# --------------------------------------------------------------------
# โจทย์ท้าทาย (Challenge): รับค่าพิกัดจากผู้ใช้ผ่าน input()
# --------------------------------------------------------------------
print("\n--- Challenge: Custom Geometric Calculation ---")

# รับค่าพิกัดและแปลงชนิดข้อมูลเป็น float เพื่อรองรับทศนิยม
x1 = float(input("กรุณากรอกพิกัด x ของจุดที่ 1: "))
y1 = float(input("กรุณากรอกพิกัด y ของจุดที่ 1: "))
x2 = float(input("กรุณากรอกพิกัด x ของจุดที่ 2: "))
y2 = float(input("กรุณากรอกพิกัด y ของจุดที่ 2: "))

# บันทึกพิกัดที่ผู้ใช้กรอกลงใน Tuple
user_point1 = (x1, y1)
user_point2 = (x2, y2)

# คำนวณระยะห่าง
user_distance = math.sqrt((user_point2[0] - user_point1[0]) ** 2 + (user_point2[1] - user_point1[1]) ** 2)

print(f"พิกัดจุดที่ 1 ที่คุณกรอก: {user_point1}")
print(f"พิกัดจุดที่ 2 ที่คุณกรอก: {user_point2}")
print(f"ระยะห่างระหว่างจุดทั้งสองคือ: {user_distance:.4f}")
```
